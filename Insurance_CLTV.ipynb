# Insurance Customer Lifetime Value Prediction

## Importing the libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Data preprocessing

```python
data = pd.read_csv('Vahan_Bima.csv')
data
```

```python
data = pd.get_dummies(data[['gender', 'area', 'qualification', 'income', 'marital_status', 'vintage', 'claim_amount', 'num_policies', 'policy', 'type_of_policy', 'cltv']], drop_first= True)
data
```

```python
X = data.drop('cltv', axis = 1)
y = data['cltv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Modeling

```python
model = LinearRegression()
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## Evaluating the model

```python
r2_score(y_test, y_pred)
```
